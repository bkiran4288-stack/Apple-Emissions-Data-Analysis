# Abschlussprojekt

## Fragestellung: Wie entwickelt sich Apples CO₂-Bilanz im Zeitraum 2015–2022 – und welche Faktoren (z. B. Produktlebenszyklus, Unternehmensstruktur, Umsatzentwicklung) haben den größten Einfluss auf Reduktion, Effizienz und Zielerreichung bis 2030?


### Ziel: Bewertung, ob Apple auf dem richtigen Weg zur Netto-Null-Emission bis 2030 ist – unter Berücksichtigung von Emissionsquellen, wirtschaftlichem Wachstum und internen Maßnahmen. Die Analyse liefert datenbasierte Handlungsempfehlungen für Unternehmen mit ähnlicher Zielsetzung.

Ermöglicht die Berechnung der Emissionseffizienz, z. B. CO₂/Produkt oder CO₂/Mitarbeiter – wichtig für Unternehmen, die wachsen, aber emissionsärmer werden wollen.
